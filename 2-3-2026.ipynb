```python
class Area:
    def __init__(self,n,m):
        self.n=n
        self.m=m
    def area(self):
        return self.n*self.m    
class triangle(Area):
    def __init__(self,n,m):
        super().__init__(n,m)

        def area(self):
            c=super().area()
            ta=c/2
            return ta
        
d = triangle (12,42)
print(d.area())
```

```python
class payment:
    def pay(self,amount):
        """""Process payment """""
        self.amount = amount
        return self.amount

class CreditCardPayment(payment):
    def pay(self, amount):
        super().pay(amount)
        print("Credit card payment of processing in Rs:", amount)

class JazzCashPayment(payment):
    def pay(self, amount):
        super().pay(amount)
        print("Jazz Cash  Payment COMPLETED in Rs:", amount)

class BankTransferPayment(payment):
    def pay(self,amount):
        super().pay(amount)
        print("Bank Transfer Successfull in Rs:",amount)

a = CreditCardPayment()
a.pay(10000)

b = JazzCashPayment()
b.pay(30000)

c = BankTransferPayment()
c.pay(400000)
```

```python
class Employee:
    def __init__(self, name):
        self.name = name

    def calculate_salary(self):
        """Return the salary for the employee. To be implemented by subclasses."""
        raise NotImplementedError("Subclasses must implement calculate_salary")


class FullTimeEmployee(Employee):
    def __init__(self, name, fixed_salary):
        super().__init__(name)
        self.fixed_salary = fixed_salary

    def calculate_salary(self):
        return self.fixed_salary


class PartTimeEmployee(Employee):
    def __init__(self, name, hourly_rate, hours_worked):
        super().__init__(name)
        self.hourly_rate = hourly_rate
        self.hours_worked = hours_worked

    def calculate_salary(self):
        return self.hourly_rate * self.hours_worked


class Intern(Employee):
    def __init__(self, name, stipend):
        super().__init__(name)
        self.stipend = stipend

    def calculate_salary(self):
        return self.stipend


# Demonstration of polymorphism
if __name__ == "__main__":
    employees = [
        FullTimeEmployee("Ali", 5000),
        PartTimeEmployee("Bilal", 20, 120),
        Intern("trump", 800),
    ]

    for emp in employees:
        print(f"{emp.name}: {emp.calculate_salary()}")
```

```python
class Customer:
    def __init__(self, name):
        self.name = name

    def get_discount(self, amount):
        """Return the discount amount for a purchase.
        Subclasses should override this to provide specific discount rates."""
        raise NotImplementedError("Subclasses must implement get_discount")


class RegularCustomer(Customer):
    DISCOUNT_RATE = 0.05

    def get_discount(self, amount):
        return amount * self.DISCOUNT_RATE


class PremiumCustomer(Customer):
    DISCOUNT_RATE = 0.10

    def get_discount(self, amount):
        return amount * self.DISCOUNT_RATE


class VIPCustomer(Customer):
    DISCOUNT_RATE = 0.20

    def get_discount(self, amount):
        return amount * self.DISCOUNT_RATE
    
class GovtSevent(Customer):
    DISCOUNT_RATE= 0.99
    
    def get_discount(self, amount):
        return amount * self.DISCOUNT_RATE

# demonstration
if __name__ == "__main__":
    amount = 1000
    customers = [
        RegularCustomer("Ahmad"),
        PremiumCustomer("Ali"),
        VIPCustomer("Hamza"),
        GovtSevent("Chiller")
    ]

    for cust in customers:
        discount = cust.get_discount(amount)
        payable = amount - discount
        print(f"{cust.name} ({cust.__class__.__name__}): discount={discount}, payable={payable}")
```

```python
class Account:
    def __init__(self, name, balance):
        self.name = name
        self.balance = balance

    def withdraw(self, amount):
        """Withdraw amount from account. Subclasses implement fee logic."""
        raise NotImplementedError("Subclasses must implement withdraw")

    def display_balance(self):
        return self.balance


class SavingsAccount(Account):
    FEE_RATE = 0.02

    def withdraw(self, amount):
        fee = amount * self.FEE_RATE
        total_deduction = amount + fee
        if total_deduction <= self.balance:
            self.balance -= total_deduction
            return f"Withdrew {amount}, Fee: {fee}, New Balance: {self.balance}"
        else:
            return "Insufficient balance"


class CurrentAccount(Account):
    FEE_RATE = 0.0

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            return f"Withdrew {amount}, Fee: 0, New Balance: {self.balance}"
        else:
            return "Insufficient balance"


class BusinessAccount(Account):
    FEE_RATE = 0.05

    def withdraw(self, amount):
        fee = amount * self.FEE_RATE
        total_deduction = amount + fee
        if total_deduction <= self.balance:
            self.balance -= total_deduction
            return f"Withdrew {amount}, Fee: {fee}, New Balance: {self.balance}"
        else:
            return "Insufficient balance"


# Demonstration
if __name__ == "__main__":
    accounts = [
        SavingsAccount("Waleed", 5000),
        CurrentAccount("Ali", 5000),
        BusinessAccount("wayne Imterprices", 50000)
    ]

    withdrawal_amount = 1000

    for acc in accounts:
        print(f"\n{acc.name} ({acc.__class__.__name__}):")
        print(acc.withdraw(withdrawal_amount))
```
