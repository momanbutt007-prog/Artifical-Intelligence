```python
# Define functions
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b
def divide(a, b):
    if b != 0:
        return a / b
    else:
        return "Cannot divide by zero"

# Show options to user
print("What do you want to do?")
print("1. Add numbers")
print("2. Subtract numbers")
print("3. Multiply numbers")
print("4. Divide numbers")

# Get user's choice
choice = input("Enter 1, 2, 3, or 4: ")

print("you choose option", choice)  

# Get numbers
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))

# Perform the chosen operation
if choice == "1":
    result = add(num1, num2)
    print(f"{num1} + {num2} = {result}")
elif choice == "2":
    result = subtract(num1, num2)
    print(f"{num1} - {num2} = {result}")
elif choice == "3":
    result = multiply(num1, num2)
    print(f"{num1} * {num2} = {result}")
elif choice == "4":
    result = divide(num1, num2)
    print(f"{num1} / {num2} = {result}")
else:
    print("Invalid choice!")
```

```python
#calculate a electricity bill based on units consumed 
units = float(input("Enter the number of units consumed: "))
if units <= 100:
    bill = units * 13
elif units <= 200:
    bill = 100 * 13 + (units - 100) * 24
else:
    bill = 100 * 13 + 100 * 24 + (units - 200) * 35

#in due date of bill is 10th of month then add 5% late fee to the bill
due_date = int(input("Enter the due date of the bill (1-31): "))

if due_date > 10:
    bill = bill + (bill * 0.05)
    print("Late fee applied")  # Add 5% late fee
else:
    print("No late fee applied.")

    #add fix GST of 17% to the bill
gst = bill + (bill * 0.17)
print(f"Your electricity bill with GST is: Rs{gst:.2f}")
print(f"Your electricity bill with late fee and GST is: Rs{gst:.2f}")
```

```python
#simple star patten
n = 5
for i in range(n):
    for j in range(i+1):
        print("*", end=" ")
    print()
for i in range(n-1, 0, -1):
    for j in range(i):
        print("*", end=" ")
    print()
```
