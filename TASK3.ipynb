# Pivot Table 
1️⃣ Average CGPA by Department

```python
df.pivot_table(values="CGPA", index=["Department", "City"], aggfunc="mean")
```

2️⃣ Final Marks by Gender

```python
df.pivot_table(values="Final_Marks", index="Gender", aggfunc="sum")
```

3️⃣ Average Attendance by Department and Gender

```python
df.pivot_table(values="Attendance_Rate", index="Department", columns="Gender", aggfunc="mean")
```

4️⃣ Total Students by Department

```python
df.pivot_table(values="Student_ID", index="Department", aggfunc="count")
```

5️⃣ Average Study Hours per Department

```python
df.pivot_table(values="Study_Hours_Per_Day", index=[ "City","Department"],columns="Semester", aggfunc="sum")
```

6️⃣ Average Midterm and Final Marks by Department

```python
df.pivot_table(values=["Midterm_Marks","Final_Marks"], index="Department", aggfunc="mean")
```

7️⃣ Fee process of each department

```python
pd.pivot_table(df, values="Student_ID" , index="Department",columns="Fees_Paid",aggfunc="count")
```

8️⃣ Pivot table to find the average Final Marks for students in each City.

```python
pd.pivot_table(df, values="Final_Marks", index="City", columns="Department",aggfunc="mean" )
```

9️⃣ Pivot table to calculate the average Attendance Rate for each City and Department with semester wise.

```python
pd.pivot_table(df, values="Attendance_Rate", index=["City","Department"], columns="Semester",aggfunc="mean" )
```
