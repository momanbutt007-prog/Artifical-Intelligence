```python
import numpy as np
import time

# Create large data
size = 100000
python_list = list(range(size))
numpy_array = np.arange(size)

# -------- Python List --------
start = time.time()
result_list = [x + 15 for x in python_list]
end = time.time()

print("Python List Time:", end - start)

# -------- NumPy Array --------
start = time.time()
result_numpy = numpy_array + 15
end = time.time()

print("NumPy Time:", end - start)
```
