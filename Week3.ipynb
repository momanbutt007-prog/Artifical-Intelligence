## 2 NumPy Array Math Wizard

```python
# Demonstrate broadcasting and universal functions.

import numpy as np

arr1 =np.array([[1,2,3],[4,5,6]])
arr2 =np.array([10,20,30])

arr3 = arr1 + arr2
print(arr3)
arr4 = arr1 - arr2
print(arr4)
arr5 = arr1 * arr2
print(arr5)
arr6 = arr1 % arr2
print(arr6)
```

```python
arr10 = np.array([1,4,9,16,25])
print(np.sqrt(arr10))
print(np.exp(arr10))
print(np.sin(arr10))
```

```python
arr11 = np.array([[5,12,3],[8,2,7]])
print(arr11)
print(arr11.sum())
print(np.max(arr11 , axis=0))
print(np.min(arr11 , axis=0))
```

## 3 Pandas DataFrame Detective

```python
# Perform basic exploratory data analysis using Pandas.

import pandas as pd

data = {
'Name': ['Ali Khan', 'Ahmed Raza', 'Usman Tariq', 'Bilal Ahmed','Hassan Ali', 'Zain Malik',],
'Age': [25,30,np.nan,35,28,np.nan],
'Score': [85,92,78,np.nan,88,95],
'Grade': ['A','A','B','C','B','A']
}

df = pd.DataFrame(data)
df
```

```python
df.head(3)
```

```python
df.shape
```

```python
df.columns
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
mean_age = df['Age'].mean()
print("Mean age:", mean_age)
```

```python
df["Age"] = df["Age"].fillna(mean_age)
df
```

```python
df.dropna(inplace=True)
df
```

```python
print(df[["Name", "Age"]])
```

```python
high_scores = df[df['Score'] > 85]
print(high_scores)
```

```python
df.set_index('Name' , inplace=True)
```

```python
df
```

```python
unique_grades = df["Grade"].unique()
print(unique_grades)
```

```python
grade_counts = df['Grade'].value_counts()
print(grade_counts)
```

```python
sorted_df = df.sort_values('Score', ascending=False)
print(sorted_df)
```
