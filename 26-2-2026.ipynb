```python
#cracker 1
import itertools
import string
import time
import threading
from queue import Queue
import concurrent.futures

class FastPasswordCracker:
    def __init__(self):
        self.found = False
        self.password = None
        self.total_attempts = 0
        self.lock = threading.Lock()
    
    def try_combinations(self, chars, length, target, start_idx, end_idx, result_queue):
        """Try combinations in a specific range"""
        local_attempts = 0
        
        # Generate only a subset of combinations
        combinations = itertools.product(chars, repeat=length)
        
        # Skip to start position (simplified - in reality you'd need more complex slicing)
        for i, guess_tuple in enumerate(combinations):
            if self.found:
                return
            
            local_attempts += 1
            
            if i >= start_idx and i < end_idx:
                guess = ''.join(guess_tuple)
                
                if guess == target:
                    with self.lock:
                        self.found = True
                        self.password = guess
                    result_queue.put((guess, local_attempts))
                    return
                
                if local_attempts % 5000 == 0:
                    print(f"Thread trying: {guess}", end='\r')
        
        with self.lock:
            self.total_attempts += local_attempts
    
    def crack_multithreaded(self, target_password, max_length):
        """Multi-threaded password cracking"""
        start_time = time.time()
        chars = string.ascii_lowercase + string.digits
        result_queue = Queue()
        
        print(f"\n🚀 Starting multi-threaded cracker for '{target_password}'...")
        
        for length in range(1, max_length + 1):
            if self.found:
                break
                
            print(f"\nTrying length {length} with multiple threads...")
            
            # Calculate total combinations for this length
            total_combos = len(chars) ** length
            num_threads = min(8, total_combos)  # Use up to 8 threads
            
            # Divide work among threads
            combos_per_thread = total_combos // num_threads
            threads = []
            
            with concurrent.futures.ThreadPoolExecutor(max_workers=num_threads) as executor:
                futures = []
                for i in range(num_threads):
                    start = i * combos_per_thread
                    end = start + combos_per_thread if i < num_threads - 1 else total_combos
                    
                    future = executor.submit(
                        self.try_combinations, 
                        chars, length, target_password, 
                        start, end, result_queue
                    )
                    futures.append(future)
                
                # Check if any thread found the password
                while not self.found and not all(f.done() for f in futures):
                    time.sleep(0.1)
                    
                    if not result_queue.empty():
                        password, attempts = result_queue.get()
                        end_time = time.time()
                        print(f"\n✅ Password cracked: '{password}'")
                        print(f"Attempts: {self.total_attempts + attempts:,}")
                        print(f"Time: {end_time - start_time:.2f} seconds")
                        return
        
        if not self.found:
            print("\n❌ Password not found within given parameters")

# Usage
cracker = FastPasswordCracker()
target = input("Enter password to test: ")
max_len = int(input("Enter max length: "))
cracker.crack_multithreaded(target, max_len)
```

```python
#cracker 2
import itertools
import string
import time

def dictionary_attack(target_password):
    """First try common passwords (fastest method)"""
    common_passwords = [
        "password", "123456", "12345678", "1234", "qwerty", 
        "abc123", "password1", "admin", "welcome", "admin123",
        "letmein", "monkey", "dragon", "baseball", "football",
        "master", "hello", "freedom", "whatever", "internet"
    ]
    
    print("\n📚 Starting dictionary attack...")
    attempts = 0
    
    for password in common_passwords:
        attempts += 1
        print(f"Trying: {password}", end='\r')
        if password == target_password:
            return True, password, attempts
    
    return False, None, attempts

def optimized_crack_password(target_password, max_length):
    """Optimized password cracker with multiple strategies"""
    start_time = time.time()
    total_attempts = 0
    
    # Strategy 1: Dictionary attack first
    found, password, attempts = dictionary_attack(target_password)
    total_attempts += attempts
    
    if found:
        end_time = time.time()
        print(f"\n✅ Password cracked using dictionary!")
        print(f"Password: '{password}'")
        print(f"Attempts: {total_attempts}")
        print(f"Time: {end_time - start_time:.2f} seconds")
        return
    
    print("\n🔍 Dictionary failed, starting brute force...")
    
    # Strategy 2: Optimized brute force
    chars = string.ascii_lowercase + string.ascii_uppercase + string.digits + "!@#$%"
    
    for length in range(1, max_length + 1):
        print(f"\nTrying length {length}...")
        
        # Use list comprehension for faster generation
        for guess_tuple in itertools.product(chars, repeat=length):
            total_attempts += 1
            guess = ''.join(guess_tuple)
            
            # Show progress every 10000 attempts
            if total_attempts % 10000 == 0:
                print(f"Attempts: {total_attempts}, Current: {guess}", end='\r')
            
            if guess == target_password:
                end_time = time.time()
                print(f"\n✅ Password cracked!")
                print(f"Password: '{guess}'")
                print(f"Attempts: {total_attempts:,}")
                print(f"Time: {end_time - start_time:.2f} seconds")
                print(f"Speed: {total_attempts/(end_time - start_time):.0f} attempts/sec")
                return

# Usage
target = input("Enter a password to test (lowercase letters/numbers recommended): ").lower()
max_len = int(input("Enter max length: "))
optimized_crack_password(target, max_len)
```

```python
#cracker 3
import itertools
import string
import time
import re

class SmartPasswordCracker:
    def __init__(self):
        self.common_patterns = [
            r'^[a-z]+$',           # all lowercase
            r'^[A-Z]+$',           # all uppercase
            r'^\d+$',              # all numbers
            r'^[a-zA-Z]+$',        # letters only
            r'^[a-z]+\d+$',         # letters then numbers
            r'^\d+[a-z]+$',         # numbers then letters
            r'^[a-zA-Z0-9]+$',      # alphanumeric
        ]
    
    def detect_pattern(self, password):
        """Detect password pattern to optimize cracking"""
        for i, pattern in enumerate(self.common_patterns):
            if re.match(pattern, password):
                return i
        return len(self.common_patterns)  # complex pattern
    
    def generate_charset(self, pattern_index):
        """Generate optimized charset based on pattern"""
        if pattern_index == 0:  # all lowercase
            return string.ascii_lowercase
        elif pattern_index == 1:  # all uppercase
            return string.ascii_uppercase
        elif pattern_index == 2:  # all numbers
            return string.digits
        elif pattern_index == 3:  # letters only
            return string.ascii_letters
        elif pattern_index == 4:  # letters then numbers
            return string.ascii_lowercase + string.digits
        elif pattern_index == 5:  # numbers then letters
            return string.digits + string.ascii_lowercase
        else:  # complex
            return string.ascii_letters + string.digits + "!@#$%"
    
    def crack_with_pattern(self, target_password):
        """Crack password using pattern detection"""
        start_time = time.time()
        
        print(f"\n🔍 Analyzing password pattern...")
        pattern_idx = self.detect_pattern(target_password)
        pattern_names = ["lowercase", "uppercase", "numbers", "letters", 
                        "letters+numbers", "numbers+letters", "complex"]
        print(f"Detected pattern: {pattern_names[pattern_idx]}")
        
        charset = self.generate_charset(pattern_idx)
        print(f"Using optimized charset ({len(charset)} characters)")
        
        attempts = 0
        length = len(target_password)
        
        print(f"\n🚀 Starting optimized brute force (length {length})...")
        
        for guess_tuple in itertools.product(charset, repeat=length):
            attempts += 1
            guess = ''.join(guess_tuple)
            
            if attempts % 10000 == 0:
                print(f"Attempts: {attempts:,} | Current: {guess}", end='\r')
            
            if guess == target_password:
                end_time = time.time()
                print(f"\n✅ Password cracked!")
                print(f"Password: '{guess}'")
                print(f"Attempts: {attempts:,}")
                print(f"Time: {end_time - start_time:.2f} seconds")
                print(f"Speed: {attempts/(end_time - start_time):.0f} attempts/sec")
                
                # Calculate search space reduction
                full_charset = string.ascii_letters + string.digits + "!@#$%"
                full_space = len(full_charset) ** length
                reduction = (1 - attempts/full_space) * 100
                print(f"Search space reduced by: {reduction:.1f}%")
                return True
        
        print("\n❌ Password not found")
        return False

# Usage
cracker = SmartPasswordCracker()
target = input("Enter password to test: ")
cracker.crack_with_pattern(target)
```
