```python
class Employee:
    def __init__(self, name, employee_id, salary, password):
        self.__name = name
        self.__employee_id = employee_id
        self.__salary = salary
        self.__password = password
        self._department = "General"
    def authenticate(self, password):
        return self.__password == password
    def display_info(self):
        print(f"Name: {self.__name}")
        print(f"ID: {self.__employee_id}")
        print(f"Salary: {self.__salary}")
        print(f"Department: {self._department}")
        print(f"Bonus: {self.calculate_bonus()}")
    def calculate_bonus(self):
        return self.__salary * 0.05
class Manager(Employee):
    def __init__(self, name, employee_id, salary, password, team_size):
        super().__init__(name, employee_id, salary, password)
        self._department = "Management"
        self.__employees = {}
        self.__manager_password = password
        self.__team_size = team_size
    def add_employee(self, employee):
        self.__employees[employee._Employee__name] = employee
    def display_all_employees(self):
        for emp in self.__employees.values():
            emp.display_info()
            print("---")
    def authenticate_manager(self, password):
        return self.__manager_password == password
    def calculate_bonus(self):
        return self._Employee__salary * 0.10 + 500 * self.__team_size
    def display_info(self):
        super().display_info()
        print(f"Team Size: {self.__team_size}")
# Create employees
employees = [
    Employee("Ali", "E1", 50000, "ali321"),
    Employee("Bilal", "E2", 60000, "bilal123"),
    Employee("Hamza", "E3", 70000, "qari123"),
    Employee("Hafsa", "G4", 80000, "gamer123"),
    Employee("Jannat", "E5", 90000, "987654"),
]
# Create manager
manager = Manager("Moman butt", "M1", 100000, "Butt1234", len(employees))
for emp in employees:
    manager.add_employee(emp)
# Login
print("1. Employee Login")
print("2. Manager Login")
choice = input("Enter choice (1/2): ")
if choice == "1":
    name = input("Enter name: ")
    password = input("Enter password: ")
    for emp in employees:
        if emp._Employee__name == name and emp.authenticate(password):
            emp.display_info()
            break
    else:
        print("Invalid credentials")
elif choice == "2":
    password = input("Enter manager password: ")
    if manager.authenticate_manager(password):
        manager.display_all_employees()
        print("\nManager Info:")
        manager.display_info()
    else:
        print("Invalid manager password")
else:
    print("Invalid choice")
```
