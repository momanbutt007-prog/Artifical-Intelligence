```python
#deepseek code
class Vehicle:
    def __init__(self, name, make_year, color):
        self.name = name
        self.make_year = make_year
        self.color = color
        self._mileage = 0  # Private attribute
        self._condition = "Excellent"  # Private attribute
    
    # Getters and Setters
    def get_name(self):
        return self.name
    
    def set_name(self, name):
        self.name = name
        print(f"✓ Vehicle name updated to: {name}")
    
    def get_make_year(self):
        return self.make_year
    
    def set_make_year(self, make_year):
        current_year = 2026
        if make_year > current_year:
            print("⚠️ That's a future year! Time traveler detected!")
        self.make_year = make_year
    
    def get_color(self):
        return self.color
    
    def set_color(self, color):
        old_color = self.color
        self.color = color
        print(f"✓ Color changed from {old_color} to {color}")
    
    def get_mileage(self):
        return self._mileage
    
    def drive(self, distance):
        """Drive the vehicle for a certain distance"""
        self._mileage += distance
        self._update_condition()
        print(f"🛣️  Drove {distance} km. Total mileage: {self._mileage} km")
    
    def _update_condition(self):
        """Update vehicle condition based on mileage"""
        if self._mileage < 50000:
            self._condition = "Excellent"
        elif self._mileage < 100000:
            self._condition = "Good"
        elif self._mileage < 150000:
            self._condition = "Fair"
        else:
            self._condition = "Needs Maintenance"
    
    def get_condition(self):
        return self._condition
    
    def display_info(self):
        print("\n" + "="*60)
        print(f"🚗 {self.name} - VEHICLE INFO")
        print("="*60)
        print(f"📝 Name: {self.name}")
        print(f"📅 Make Year: {self.make_year}")
        print(f"🎨 Color: {self.color}")
        print(f"📊 Mileage: {self._mileage} km")
        print(f"🔧 Condition: {self._condition}")
        print("-"*60)

class Car(Vehicle):
    def __init__(self, brand, model, price, color, car_type, make_year):
        super().__init__(brand, make_year, color)
        self.model = model
        self.price = price
        self.car_type = car_type
        self._owner = "Dealership"
        self._is_sold = False
        self._fuel_level = 100  # Percentage
        self._features = []
    
    # Getters and Setters
    def get_model(self):
        return self.model
    
    def set_model(self, model):
        self.model = model
        print(f"✓ Model updated to: {model}")
    
    def get_price(self):
        return self.price
    
    def set_price(self, price):
        old_price = self.price
        self.price = price
        if price > old_price:
            print(f"📈 Price increased from ${old_price:,} to ${price:,}")
        else:
            print(f"📉 Price decreased from ${old_price:,} to ${price:,}")
    
    def get_car_type(self):
        return self.car_type
    
    def set_car_type(self, car_type):
        self.car_type = car_type
        print(f"✓ Car type changed to: {car_type}")
    
    def get_owner(self):
        return self._owner
    
    def sell_car(self, new_owner):
        """Sell the car to a new owner"""
        if not self._is_sold:
            self._owner = new_owner
            self._is_sold = True
            print(f"\n🎉 SOLD! {self.get_full_name()} has been sold to {new_owner}!")
            print(f"💰 Final sale price: ${self.price:,}")
        else:
            print(f"❌ {self.get_full_name()} has already been sold!")
    
    def add_feature(self, feature):
        """Add a feature to the car"""
        self._features.append(feature)
        print(f"✅ Added feature: {feature}")
    
    def list_features(self):
        """Display all car features"""
        if self._features:
            print("\n✨ Car Features:")
            for i, feature in enumerate(self._features, 1):
                print(f"   {i}. {feature}")
        else:
            print("📦 No special features added yet.")
    
    def refuel(self, amount):
        """Refuel the car"""
        if self._fuel_level + amount <= 100:
            self._fuel_level += amount
            print(f"⛽ Refueled! Fuel level: {self._fuel_level}%")
        else:
            self._fuel_level = 100
            print(f"⛽ Tank is full! Fuel level: 100%")
    
    def drive(self, distance):
        """Override drive method for car"""
        super().drive(distance)
        # Consume fuel (rough calculation)
        fuel_consumed = distance / 10  # Assuming 10 km per 1% fuel
        self._fuel_level = max(0, self._fuel_level - fuel_consumed)
        print(f"⛽ Fuel level: {self._fuel_level:.1f}%")
        
        if self._fuel_level < 10:
            print("⚠️  Low fuel! Time to refuel!")
    
    def get_full_name(self):
        """Get complete car name"""
        return f"{self.name} {self.model}"
    
    def apply_discount(self, discount_percent):
        """Apply discount to car price"""
        discount_amount = self.price * (discount_percent / 100)
        self.price -= discount_amount
        print(f"💰 {discount_percent}% discount applied!")
        print(f"   New price: ${self.price:,.2f} (You save: ${discount_amount:,.2f})")
        return self.price
    
    def age_calculation(self, current_year=2026):
        """Calculate age of the car"""
        age = current_year - self.make_year
        if age == 0:
            return "Brand New (Current Year)"
        elif age == 1:
            return f"{age} year old"
        else:
            return f"{age} years old"
    
    def compare_price(self, other_car):
        """Compare price with another car"""
        print(f"\n🔍 Price Comparison:")
        print(f"   {self.get_full_name()}: ${self.price:,}")
        print(f"   {other_car.get_full_name()}: ${other_car.price:,}")
        
        if self.price > other_car.price:
            difference = self.price - other_car.price
            print(f"   {self.get_full_name()} is ${difference:,} more expensive")
        elif self.price < other_car.price:
            difference = other_car.price - self.price
            print(f"   {other_car.get_full_name()} is ${difference:,} more expensive")
        else:
            print(f"   Both cars have the same price!")
    
    def display_info(self):
        """Override display method with more details"""
        super().display_info()
        print(f"🏷️  Brand: {self.name}")
        print(f"🚘 Model: {self.model}")
        print(f"💰 Price: ${self.price:,}")
        print(f"🏎️  Type: {self.car_type}")
        print(f"👤 Owner: {self._owner}")
        print(f"⛽ Fuel: {self._fuel_level:.1f}%")
        print(f"📅 Age: {self.age_calculation()}")
        self.list_features()
        print("="*60)

# Create some exciting cars with interesting features
print("\n" + "🔥"*30)
print("     WELCOME TO THE DOGAR MOTORS SHOWROOM")
print("🔥"*30)

# Create car objects
cars = [
    Car("Toyota", "Camry", 28500, "Silver", "Sedan", 2024),
    Car("Honda", "Civic", 25000, "Red", "Sedan", 2022),
    Car("Rolls Royce", "Phantom", 4500000, "Maroon", "Luxury", 2024),
    Car("Bugatti", "Chiron", 5000000, "Coppery", "Hyper Sport", 2025),
    Car("Mercedes", "G-Class", 130000, "Black", "Luxury SUV", 2023),
    Car("Ferrari", "SF90 Stradale", 625000, "Rosso Corsa", "Hybrid Supercar", 2024),
    Car("Lamborghini", "Urus", 230000, "Orange", "Super SUV", 2023)
]

# Add features to some cars
cars[2].add_feature("Starlight Headliner")
cars[2].add_feature("Champagne Cooler")
cars[2].add_feature("Massaging Seats")

cars[3].add_feature("Quad Turbo W16 Engine")
cars[3].add_feature("0-60 in 2.3 seconds")
cars[3].add_feature("Top Speed 490 km/h")

cars[5].add_feature("1000 HP Hybrid V8")
cars[5].add_feature("F1-derived Tech")
cars[5].add_feature("Carbon Fiber Body")

# Interactive Showroom Experience
def showroom_experience():
    print("\n" + "="*70)
    print("🏁  WELCOME TO THE DOGAR MOTORS  🏁")
    print("="*70)
    
    while True:
        print("\n📋 MENU:")
        print("1. View All Cars")
        print("2. View Luxury Cars (Price > $1M)")
        print("3. View Sports Cars")
        print("4. Compare Two Cars")
        print("5. Test Drive a Car")
        print("6. Buy a Car")
        print("7. Exit")
        
        choice = input("\nEnter your choice (1-7): ")
        
        if choice == '1':
            for i, car in enumerate(cars, 1):
                print(f"\n--- Car #{i} ---")
                car.display_info()
        
        elif choice == '2':
            luxury_cars = [car for car in cars if car.price > 1000000]
            if luxury_cars:
                print("\n✨ LUXURY CARS (Over $1M):")
                for car in luxury_cars:
                    print(f"   • {car.get_full_name()} - ${car.price:,}")
            else:
                print("No luxury cars found.")
        
        elif choice == '3':
            sports_cars = [car for car in cars if "Sport" in car.car_type or "Super" in car.car_type]
            if sports_cars:
                print("\n🏎️  SPORTS CARS:")
                for car in sports_cars:
                    print(f"   • {car.get_full_name()} - {car.car_type}")
            else:
                print("No sports cars found.")
        
        elif choice == '4':
            print("\nAvailable Cars:")
            for i, car in enumerate(cars, 1):
                print(f"{i}. {car.get_full_name()}")
            
            try:
                idx1 = int(input("\nEnter first car number: ")) - 1
                idx2 = int(input("Enter second car number: ")) - 1
                
                if 0 <= idx1 < len(cars) and 0 <= idx2 < len(cars):
                    cars[idx1].compare_price(cars[idx2])
                else:
                    print("Invalid car numbers!")
            except ValueError:
                print("Please enter valid numbers!")
        
        elif choice == '5':
            print("\nAvailable for Test Drive:")
            for i, car in enumerate(cars, 1):
                print(f"{i}. {car.get_full_name()}")
            
            try:
                idx = int(input("Which car would you like to test drive? ")) - 1
                if 0 <= idx < len(cars):
                    car = cars[idx]
                    print(f"\n🚗 Test Driving {car.get_full_name()}!")
                    distance = int(input("How many km would you like to drive? "))
                    car.drive(distance)
                else:
                    print("Invalid car number!")
            except ValueError:
                print("Please enter a valid number!")
        
        elif choice == '6':
            print("\n🏁 READY TO BUY?")
            for i, car in enumerate(cars, 1):
                print(f"{i}. {car.get_full_name()} - ${car.price:,}")
            
            try:
                idx = int(input("Which car would you like to buy? ")) - 1
                if 0 <= idx < len(cars):
                    car = cars[idx]
                    name = input("Enter your name: ")
                    
                    # Negotiation simulation
                    print(f"\n🤝 Negotiating price for {car.get_full_name()}...")
                    negotiate = input("Would you like to negotiate? (yes/no): ")
                    
                    if negotiate.lower() == 'yes':
                        discount = float(input("What discount percentage do you want? "))
                        if discount > 0:
                            car.apply_discount(min(discount, 15))  # Max 15% discount
                    
                    car.sell_car(name)
                else:
                    print("Invalid car number!")
            except ValueError:
                print("Please enter a valid number!")
        
        elif choice == '7':
            print("\n👋 Thanks for visiting our showroom! Drive safe!")
            break
        
        else:
            print("❌ Invalid choice! Please try again.")

# Run the showroom experience
if __name__ == "__main__":
    showroom_experience()

# Also show the original cars
print("\n" + "⭐"*30)
print("     FEATURED CARS IN OUR COLLECTION")
print("⭐"*30)

for i, car in enumerate(cars, 1):
    print(f"\n🌟 Car #{i}: {car.get_full_name()}")
    car.display_info()
```
