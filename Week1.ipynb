## 1 The Personal Assistant 

```python
# Step 1: Ask for user's full name
user_name = input("Enter your full name: ")

# Step 2: Ask for year of birth
dob_year = int(input("Enter your birth year (e.g. 2000): "))

# Step 3: Ask for current age
current_age = int(input("Enter your age right now: "))

# Step 4: Ask for present year
year_now = int(input("What is the current year? "))

# Step 5: Calculate 100 years milestone
hundred_year = year_now + (100 - current_age)

# Step 6: Ask for favorite number
fav_num = float(input("Enter your favorite number: "))

# Step 7: Display results
print("\n----- USER SUMMARY -----")
print(f"Welcome {user_name}!")
print(f"Birth Year: {dob_year}")
print(f"Current Age: {current_age}")
print(f"Favorite Number: {fav_num}")
print(f"You will turn 100 in: {hundred_year}")
```

## 2 The Shopping List Manager 

```python
#  Build a program that manages a shopping list using lists and dictionaries.

items = ["rice", "tea", "sugar", "oranges", "meat"]
prices = {"rice": 250, "tea": 900, "sugar": 150, "oranges": 300}

print("\nInitial Items:", items)
print("Item at index 2:", items[2])

# Adding items
items.append("butter")
items.append("biscuits")
print("Added butter and biscuits")

# Removing item
items.remove("meat")
print("Removed meat")

print("Updated Items:", items)

# Price lookup
print("Prices Dictionary:", prices)
print("Price of oranges:", prices["oranges"], "Rs")

print("\n----- FINAL LIST -----")
print("Items to purchase:", items)
print("Available prices:", prices)
```

# -------- FizzBuzz Program --------

```python

print("\nFizzBuzz from 1 to 20:\n")

for num in range(1, 21):
    if num % 3 == 0 and num % 5 == 0:
        print("FizzBuzz")
    elif num % 3 == 0:
        print("Fizz")
    elif num % 5 == 0:
        print("Buzz")
    else:
        print(num)
```

# -------- Countdown Program --------

```python
count_num = int(input("\nEnter a number to start countdown: "))

if count_num > 10:
    print(f"\n{count_num} is too large. Countdown stopped!")
else:
    print(f"\nStarting countdown from {count_num}...")
    while count_num >= 1:
        print("Counting:", count_num)
        count_num -= 1
    print("🚀 Blast Off!")
```
