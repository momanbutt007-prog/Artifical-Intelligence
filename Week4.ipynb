## 2 Data Visualization with Pandas and Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("Sales.csv")
df
```

#  - Line plot of product sales over months


```python
sns.lineplot(data=df, x="product", y="price")
```

# - Bar plot of product_a by month

```python
sns.barplot(data=df , x="product", y="sales")
```

# - Pie chart of total sales by region

```python
df.groupby('region')['sales'].sum().plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8), title='Total Sales by region', colors=['gold', 'lightcoral', 'lightskyblue', 'lightgreen']); plt.ylabel(''); plt.show()
```

# 4. Seaborn Distribution Plots:

   - distplot for product_a


```python
sns.histplot(data=df,x="product",kde=True,bins=30)
```


   - jointplot comparing product_a and product_b


```python
sns.jointplot(data=df,x="product",y="price")
```

# 5. Seaborn Categorical Plots:
   - boxplot for product_a by region


```python
sns.boxplot(data=df,x="region",y="product")
```

 - barplot showing average product sales by region


```python
sns.barplot(data=df,x="product",y="price",estimator=np.mean)
```

# 6. Matrix Plot:
   - Create correlation matrix of product columns


```python
# Pivot table: products as columns
product_pivot = df.pivot_table(
    values='sales',
    index='order_id',
    columns='product',
    fill_value=0
)

# Correlation matrix
corr_matrix = product_pivot.corr()

# Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')

plt.title("Product-wise Correlation Matrix")
plt.xlabel("Products")
plt.ylabel("Products")

plt.show()
```
