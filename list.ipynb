```python
import itertools
import string
import time

def crack_password(target_password, max_length):
    chars = string.ascii_lowercase + string.digits
    attempts = 0
    start_time = time.time()

    for length in range(1, max_length + 1):
        for guess in itertools.product(chars, repeat=length):
            attempts += 1
            guess_password = ''.join(guess)
            print(f"Trying: {guess_password}", end='\r')
            if guess_password == target_password:
                end_time = time.time()
                print(f"\nPassword cracked! '{guess_password}' in {attempts} attempts")
                print(f"Time taken: {end_time - start_time} seconds")
                return

# Example usage
target_password = input("Enter a password to crack: ")
max_length = int(input("Enter the maximum password length: "))
crack_password(target_password, max_length)
```
