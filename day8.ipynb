```python
##create a class called person with name and age and method introduce that prints the name and age of the person.
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def introduce(self):
        print(f"hi my name is {self.name} and my age is {self.age}")

# Test the class
p1 = Person("hafsa", 21)
p1.introduce()
p2 = Person("Sarah", 30)
p2.introduce()
p3 = Person("momin", 25)
p3.introduce()

```

```python
## CREATE A CLASS CALLED RECTANGLE WITH LENGTH AND WIDTH AND METHODS AREA AND PERIMETER.
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width
    
    def area(self):
        return self.length * self.width
    
    def perimeter(self):
        return 2 * (self.length + self.width)

# Test the class
r1 = Rectangle(5, 3)
print(f"Area: {r1.area()}")
print(f"Perimeter: {r1.perimeter()}")

```

```python
## create a class called circle with  radius and methods circumference.
class Circle:
    def __init__(self, radius):
        self.radius = radius
    def circumference(self):
        return 2 * 3.14 * self.radius
# Test the class
c1 = Circle(67)
print(f"Circumference: {c1.circumference()}")
```

```python
##create a class student attribute marks (dictionary of subject and marks) and method average that calculates the average marks of the student.
class Student:
    def __init__(self, marks):
        self.marks = marks
    
    def average(self):
        total_marks = sum(self.marks.values())
        count = len(self.marks)
        return total_marks / count
# Test the class
marks = {"Math": 85, 
        "Science": 90, 
        "English": 78}
s1 = Student(marks)
print(f"Average Marks: {s1.average()}")
```

```python
## Create a class Car that constructs with brand and speed and has a method show_details that prints brand and speed.
class Car:
    def __init__(self, brand, speed):
        self.brand = brand
        self.speed = speed
    
    def show_details(self):
        print(f"Brand: {self.brand}, Speed: {self.speed} km/h")

# Create one object and display details
car1 = Car("civic", 120)
car1.show_details()
```

```python
## Create a class Book that takes title and price and has a method discount that returns price after 10% discount.
class Book:
    def __init__(self, title, price):
        self.title = title
        self.price = price
    
    def discount(self):
        return self.price * 0.9

# Create one object and print discounted price
book1 = Book("Python Basics", 500)
print(f"Title: {book1.title}")
print(f"Original Price: {book1.price}")
print(f"Discounted Price (10% off): {book1.discount()}")
```
