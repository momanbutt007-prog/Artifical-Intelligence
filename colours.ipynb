```python
# ANSI color codes
class Colors:
    # Text colors
    BLACK = '\033[30m'
    RED = '\033[31m'
    GREEN = '\033[32m'
    YELLOW = '\033[33m'
    BLUE = '\033[34m'
    MAGENTA = '\033[35m'
    CYAN = '\033[36m'
    WHITE = '\033[37m'
    
    # Bright colors
    BRIGHT_BLACK = '\033[90m'
    BRIGHT_RED = '\033[91m'
    BRIGHT_GREEN = '\033[92m'
    BRIGHT_YELLOW = '\033[93m'
    BRIGHT_BLUE = '\033[94m'
    BRIGHT_MAGENTA = '\033[95m'
    BRIGHT_CYAN = '\033[96m'
    BRIGHT_WHITE = '\033[97m'
    
    # Background colors
    BG_BLACK = '\033[40m'
    BG_RED = '\033[41m'
    BG_GREEN = '\033[42m'
    BG_YELLOW = '\033[43m'
    BG_BLUE = '\033[44m'
    BG_MAGENTA = '\033[45m'
    BG_CYAN = '\033[46m'
    BG_WHITE = '\033[47m'
    
    # Styles
    BOLD = '\033[1m'
    UNDERLINE = '\033[4m'
    RESET = '\033[0m'  # Resets all formatting

# Simple function to print in color
def print_color(text, color):
    print(f"{color}{text}{Colors.RESET}")

# Examples
print_color("This is RED text", Colors.RED)
print_color("This is GREEN text", Colors.GREEN)
print_color("This is BLUE text", Colors.BLUE)
print_color("This is BOLD text", Colors.BOLD)
print_color("This is UNDERLINED text", Colors.UNDERLINE)

# With background
print(f"{Colors.BG_YELLOW}{Colors.BLACK}Black text on yellow background{Colors.RESET}")
```

```python
def print_colored(text, color_name):
    colors = {
        'red': '\033[91m',
        'green': '\033[92m',
        'yellow': '\033[93m',
        'blue': '\033[94m',
        'magenta': '\033[95m',
        'cyan': '\033[96m',
        'white': '\033[97m',
        'black': '\033[90m',
        'reset': '\033[0m'
    }
    
    color_code = colors.get(color_name.lower(), colors['reset'])
    print(f"{color_code}{text}{colors['reset']}")

# Example usage
print_colored("Hello World!", "red")
print_colored("Python is awesome!", "green")
print_colored("This is blue text", "blue")
print_colored("Warning message", "yellow")

# Get color from user
user_text = input("Enter text to print: ")
user_color = input("Enter color (red/green/blue/yellow): ")
print_colored(user_text, user_color)
```

```python
def color_printer():
    colors = {
        '1': {'name': 'Red', 'code': '\033[91m'},
        '2': {'name': 'Green', 'code': '\033[92m'},
        '3': {'name': 'Yellow', 'code': '\033[93m'},
        '4': {'name': 'Blue', 'code': '\033[94m'},
        '5': {'name': 'Magenta', 'code': '\033[95m'},
        '6': {'name': 'Cyan', 'code': '\033[96m'},
        '7': {'name': 'White', 'code': '\033[97m'},
        '8': {'name': 'Black', 'code': '\033[90m'}
    }
    reset = '\033[0m'
    
    print("=== COLOR PRINTER ===")
    print("Available colors:")
    for key, color in colors.items():
        print(f"{key}. {color['name']}")
    
    # Get text
    text = input("\nEnter your text: ")
    
    # Get color choice
    choice = input("Choose color number: ")
    
    if choice in colors:
        color_code = colors[choice]['code']
        print(f"\n{color_code}{text}{reset}")
    else:
        print("Invalid color choice!")

# Run the color printer
color_printer()
```
