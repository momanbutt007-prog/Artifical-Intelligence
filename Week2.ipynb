## 1 The Data Analyzer

```python
import math

# -------- Statistical Functions --------

def get_numbers():
    nums = input("Enter values (separated by space): ")
    return [float(n) for n in nums.split()]

def calculate_mean(data):
    return sum(data) / len(data)

def calculate_variance(data, mean):
    squared_diff_sum = 0
    for value in data:
        squared_diff_sum += (value - mean) ** 2
    return squared_diff_sum / (len(data) - 1)

def calculate_std_dev(variance):
    return math.sqrt(variance)


# -------- Main Program --------

numbers = get_numbers()

if len(numbers) == 0:
    print("No input provided.")

elif len(numbers) == 1:
    mean = calculate_mean(numbers)

    print("\n--- RESULT ---")
    print(f"Numbers: {numbers}")
    print(f"Mean: {mean:.2f}")

else:
    mean = calculate_mean(numbers)
    variance = calculate_variance(numbers, mean)
    std_dev = calculate_std_dev(variance)

    print("\n--- RESULT ---")
    print(f"Numbers: {numbers}")
    print(f"Mean: {mean:.2f}")
    print(f"Variance: {variance:.2f}")
    print(f"Standard Deviation: {std_dev:.2f}")
    print("=" * 40)
```

## 2 The Student Class System

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display_info(self):
        print(f"Name: {self.name} | Age: {self.age}")


class Student(Person):
    def __init__(self, name, age, student_id, major):
        super().__init__(name, age)
        self.student_id = student_id
        self.major = major

    def display_info(self):
        super().display_info()
        print(f"Student ID: {self.student_id} | Major: {self.major}")


class GraduateStudent(Student):
    def __init__(self, name, age, student_id, major, thesis):
        super().__init__(name, age, student_id, major)
        self.thesis = thesis

    def display_info(self):
        super().display_info()
        print(f"Thesis Topic: {self.thesis}")


# Objects
p = Person("Ali Khan", 28)
s = Student("Sara Ahmed", 21, "ST101", "Software Engineering")
g = GraduateStudent("Usman Ali", 25, "GR202", "AI", "Machine Learning Models")

print("\n--- PERSON ---")
p.display_info()

print("\n--- STUDENT ---")
s.display_info()

print("\n--- GRADUATE STUDENT ---")
g.display_info()

```

## 3 Shape Calculator with Magic Methods

```python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def __str__(self):
        return f"Rectangle({self.width} x {self.height})"

    def area(self):
        return self.width * self.height

    def __add__(self, other):
        if isinstance(other, Rectangle):
            return Rectangle(self.width + other.width,
                             self.height + other.height)
        return NotImplemented

    def __eq__(self, other):
        if isinstance(other, Rectangle):
            return self.area() == other.area()
        return False


# Objects
r1 = Rectangle(5, 6)
r2 = Rectangle(3, 4)
r3 = Rectangle(2, 6)

print("\n--- RECTANGLES ---")
print("Rectangle 1:", r1)
print("Rectangle 2:", r2)

print("Area r1:", r1.area())

r_sum = r1 + r2
print("Combined Rectangle:", r_sum)

print("r1 == r2 ?", r1 == r2)
print("r2 == r3 ?", r2 == r3)
```
