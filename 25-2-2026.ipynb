```python
#deep seek
class Restaurant:
    def __init__(self, name="My Restaurant"):
        self.name = name
        self.menu = {}
        self.admin_password = "admin321"
        self.admin_password = input("Enter your passward")  
    
    def admin_login(self):
        """Simple admin authentication"""
        password = input("Enter admin password: ")
        return password == self.admin_password
    
    def add_item(self, item_name, price):
        """Add a new item to the menu"""
        try:
            price = float(price)
            if price <= 0:
                print("Price must be positive!")
                return
            self.menu[item_name.title()] = price
            print(f"✓ Item '{item_name.title()}' added to the menu with price Rs {price:.2f}.")
        except ValueError:
            print("Invalid price! Please enter a number.")
    
    def update_item(self, item_name, new_price):
        """Update an existing item's price"""
        item_name = item_name.title()
        try:
            new_price = float(new_price)
            if new_price <= 0:
                print("Price must be positive!")
                return
            if item_name in self.menu:
                old_price = self.menu[item_name]
                self.menu[item_name] = new_price
                print(f"✓ Item '{item_name}' updated from Rs {old_price:.2f} to Rs {new_price:.2f}.")
            else:
                print(f"✗ Item '{item_name}' not found in the menu.")
        except ValueError:
            print("Invalid price! Please enter a number.")
    
    def delete_item(self, item_name):
        """Delete an item from the menu"""
        item_name = item_name.title()
        if item_name in self.menu:
            del self.menu[item_name]
            print(f"✓ Item '{item_name}' deleted from the menu.")
        else:
            print(f"✗ Item '{item_name}' not found in the menu.")
    
    def view_menu(self):
        """Display the current menu"""
        print(f"\n{'='*40}")
        print(f"{self.name} - MENU")
        print('='*40)
        if self.menu:
            # Sort menu items alphabetically
            sorted_items = sorted(self.menu.items())
            for item, price in sorted_items:
                print(f"{item:<20} Rs {price:>8.2f}")
            
            # Show total items and average price
            print('-'*40)
            print(f"Total Items: {len(self.menu)}")
            avg_price = sum(self.menu.values()) / len(self.menu)
            print(f"Average Price: Rs {avg_price:.2f}")
        else:
            print("The menu is currently empty.")
        print('='*40)
    
    def search_item(self, item_name):
        """Search for an item in the menu"""
        item_name = item_name.title()
        if item_name in self.menu:
            print(f"Found: {item_name} - Rs {self.menu[item_name]:.2f}")
        else:
            # Check for partial matches
            matches = [item for item in self.menu.keys() 
                      if item_name.lower() in item.lower()]
            if matches:
                print(f"Did you mean:")
                for match in matches:
                    print(f"  {match} - Rs {self.menu[match]:.2f}")
            else:
                print(f"Item '{item_name}' not found.")
    
    def clear_menu(self):
        """Clear all items from menu (admin only)"""
        if self.admin_login():
            confirm = input("Are you sure you want to clear all items? (yes/no): ")
            if confirm.lower() == 'yes':
                self.menu.clear()
                print("Menu cleared successfully!")
            else:
                print("Operation cancelled.")
        else:
            print("Unauthorized access!")

# Admin Interface
def admin_interface():
    restaurant = Restaurant("KASHMIR POINT RESTURENT")
    
    # Add some initial items
    initial_items = [
        ("Pizza", 1800),
        ("Burger", 600),
        ("Pasta", 1000),
        ("Fries", 500),
        ("Drinks", 200),
        ("Brownies", 400),
        ("Sundae", 300)
    ]
    
    for item, price in initial_items:
        restaurant.add_item(item, price)
    
    while True:
        print(f"\n{restaurant.name} - Admin Panel")
        print("1. View Menu")
        print("2. Add Item")
        print("3. Update Item")
        print("4. Delete Item")
        print("5. Search Item")
        print("6. Clear Menu (Admin)")
        print("7. Exit '\033[31m'Admin PaneL '\033[31m'")
        
        choice = input("\nEnter your choice (1-7): ")
        
        if choice == '1':
            restaurant.view_menu()
        
        elif choice == '2':
            item = input("Enter item name: ")
            price = input("Enter item price: Rs ")
            restaurant.add_item(item, price)
        
        elif choice == '3':
            item = input("Enter item name to update: ")
            price = input("Enter new price: Rs ")
            restaurant.update_item(item, price)
        
        elif choice == '4':
            item = input("Enter item name to delete: ")
            restaurant.delete_item(item)
        
        elif choice == '5':
            item = input("Enter item name to search: ")
            restaurant.search_item(item)
        
        elif choice == '6':
            restaurant.clear_menu()
        
        elif choice == '7':
            print("Exiting Admin Panel...")
            break
        
        else:
            print("Invalid choice! Please try again.")

# Customer Interface
def customer_interface():
    restaurant = Restaurant("Foodie's Paradise")
    
    # Add some initial items
    initial_items = [
        ("Pizza", 1800),
        ("Burger", 600),
        ("Pasta", 1000),
        ("Fries", 500),
        ("Drinks", 200),
        ("Brownies", 400),
        ("Sundae", 300)
    ]
    
    for item, price in initial_items:
        restaurant.add_item(item, price)
    
    while True:
        print(f"\n{restaurant.name} - Customer View")
        print("1. View Menu")
        print("2. Search Item")
        print("3. Exit")
        
        choice = input("\nEnter your choice (1-3): ")
        
        if choice == '1':
            restaurant.view_menu()
        
        elif choice == '2':
            item = input("Enter item name to search: ")
            restaurant.search_item(item)
        
        elif choice == '3':
            print("Thank you for visiting!")
            break
        
        else:
            print("Invalid choice! Please try again.")

# Main program
def main():
    while True:
        print("\n" + "="*40)
        print(" RESTAURANT MANAGEMENT SYSTEM ")
        print("="*40)
        print("1.Admin Panel")
        print("2. Customer View ")
        print("3. Exit ")
        
        choice = input("\nSelect mode (1-3): ")
        
        if choice == '1':
            admin_interface()
        elif choice == '2':
            customer_interface()
        elif choice == '3':
            print("Goodbye!")
            break
        else:
            print("Invalid choice!")

# Run the program
if __name__ == "__main__":
    main()
```

```python
#grok
from datetime import datetime
import os

class Restaurant:
    def __init__(self, name="Spicy Haven"):
        self.name = name
        self.menu = {}
        self.categories = {}          # e.g. "Pizza", "Burgers", "Desserts", "Drinks"
        self.last_updated = datetime.now().strftime("%Y-%m-%d %H:%M")
        self.currency = "Rs"

    def clear_screen(self):
        """Cross-platform clear terminal"""
        os.system('cls' if os.name == 'nt' else 'clear')

    def add_item(self, name, price, category="Main"):
        name = name.strip().title()
        if name in self.menu:
            print(f"→ '{name}' already exists! Use update instead.")
            return
        
        self.menu[name] = {
            "price": float(price),
            "category": category.title(),
            "added_on": datetime.now().strftime("%Y-%m-%d")
        }
        
        # Auto-create category if new
        if category.title() not in self.categories:
            self.categories[category.title()] = []
        self.categories[category.title()].append(name)
        
        print(f"✓ Added: {name} → {self.currency} {price:,}  ({category})")

    def update_price(self, name, new_price):
        name = name.strip().title()
        if name in self.menu:
            old_price = self.menu[name]["price"]
            self.menu[name]["price"] = float(new_price)
            self.menu[name]["added_on"] = datetime.now().strftime("%Y-%m-%d")  # treat as updated
            print(f"→ Updated: {name}  {self.currency} {old_price:,} → {self.currency} {new_price:,}")
        else:
            print(f"✗ Item '{name}' not found.")

    def delete_item(self, name):
        name = name.strip().title()
        if name in self.menu:
            category = self.menu[name]["category"]
            del self.menu[name]
            # Clean up category list if needed
            if category in self.categories:
                self.categories[category] = [item for item in self.categories[category] if item != name]
                if not self.categories[category]:
                    del self.categories[category]
            print(f"🗑  Removed: {name}")
        else:
            print(f"✗ '{name}' not found in menu.")

    def show_menu(self, by_category=False):
        self.clear_screen()
        print(f"\n╔════════════════════════════════════════════╗")
        print(f"║          {self.name.upper()} MENU                  ║")
        print(f"╠════════════════════════════════════════════╣")
        print(f"║  Last updated: {self.last_updated}         ║")
        print(f"╚════════════════════════════════════════════╝\n")

        if not self.menu:
            print("         🍽️  Menu is empty right now...    \n")
            return

        if by_category and self.categories:
            for cat, items in sorted(self.categories.items()):
                if not items:
                    continue
                print(f"  ┌─ {cat.upper()} ─{'─' * (30 - len(cat))}")
                for item in sorted(items):
                    price = self.menu[item]["price"]
                    print(f"  │  {item:<22} {self.currency} {price:>6,.0f}")
                print("  └" + "─" * 36 + "\n")
        else:
            # Simple alphabetical list
            for item in sorted(self.menu.keys()):
                price = self.menu[item]["price"]
                cat = self.menu[item]["category"]
                print(f"  {item:<22} {self.currency} {price:>6,.0f}  ({cat})")

        print(f"\nTotal items: {len(self.menu):>3}\n")

    def admin_menu(self):
        while True:
            self.clear_screen()
            print(f"\n=== {self.name} Admin Panel ===\n")
            print("1. Add new item")
            print("2. Update price")
            print("3. Delete item")
            print("4. View full menu (simple)")
            print("5. View menu by category")
            print("6. Exit")
            choice = input("\n→ Choose (1-6): ").strip()

            if choice == "1":
                name = input("Item name: ").strip()
                if not name:
                    continue
                price = input("Price: ").strip()
                cat = input("Category (Main, Sides, Drinks, Dessert...): ").strip() or "Main"
                try:
                    self.add_item(name, price, cat)
                except ValueError:
                    print("Invalid price! Use numbers only.")
                input("\nPress Enter to continue...")

            elif choice == "2":
                name = input("Item to update: ").strip()
                price = input("New price: ").strip()
                try:
                    self.update_price(name, price)
                except ValueError:
                    print("Invalid price!")
                input("\nPress Enter...")

            elif choice == "3":
                name = input("Item to delete: ").strip()
                if input(f"Confirm delete '{name}'? (y/n): ").lower().startswith('y'):
                    self.delete_item(name)
                input("\nPress Enter...")

            elif choice == "4":
                self.show_menu(by_category=False)
                input("\nPress Enter...")

            elif choice == "5":
                self.show_menu(by_category=True)
                input("\nPress Enter...")

            elif choice in ("6", "exit", "q"):
                print("\nGoodbye! 🍕\n")
                break

            else:
                print("Invalid choice!")
                input("\nPress Enter...")


# ────────────────────────────────────────────────
if __name__ == "__main__":
    restaurant = Restaurant("Foodie Fiesta")

    # Adding some initial items (you can remove this)
    initial_items = [
        ("Margherita Pizza", 1899, "Pizza"),
        ("Pepperoni Pizza", 2299, "Pizza"),
        ("Zinger Burger", 799, "Burgers"),
        ("Cheese Fries", 599, "Sides"),
        ("Crispy Wings (6 pcs)", 950, "Sides"),
        ("Chocolate Brownie Sundae", 650, "Desserts"),
        ("Classic Cola", 250, "Drinks"),
        ("Mint Mojito", 399, "Drinks"),
    ]

    for name, price, cat in initial_items:
        restaurant.add_item(name, price, cat)

    # Start admin panel
    restaurant.admin_menu()
```
