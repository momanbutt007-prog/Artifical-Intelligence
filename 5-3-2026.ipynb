```python
import numpy as np
```

```python
x= np.array([[1,2,3,4,5],[11,22,33,44,55],[9,8,7,6,5],[99,88,77,66,55]])
print(x)
```

```python
x.ndim
x.dtype
x.argmax()
x.argmin()
x.max
x.min
x.size
x.flatten()
x.reshape(5,4)
print("Dimention of array is :", x.ndim)
print("dtype of array is :", x.dtype)
print("max number index is :" ,x.argmax())
print("Min number inder is :", x.argmin())
print("max number of array is:",x.max())
print("min number of array is :",x.min())
print("the size of array is :",x.size)
print("changing array in 1D",x.flatten())
print("After reshaping the new array is ")
print(x.reshape(5,4))

```

```python
np.mean(x)
np.median(x)
print("mean of array is :",np.mean(x))
print("median of array is :",np.median(x))
```

```python
import numpy as np
from scipy import stats
mode = stats.mode(x)
print("mode of array is :",mode)
```

```python
df_null = pd.DataFrame({
    'A': [1, 2, np.nan, 4, 5, np.nan],
    'B': [np.nan, 2, 3, np.nan, 5, 9],
    'C': [1, 2, 3, 4, 5, np.nan]
})

# Check for nulls
print(df_null.isnull())
print("---------------------")
print(df_null.isnull().sum())
print("---------------------")
print(df_null.isnull().sum().sum())
print("---------------------")
# Check for non-nulls
print(df_null.notnull())
print("---------------------")
print(df_null.notnull().sum())
print("---------------------")
```
